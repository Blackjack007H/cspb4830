```python
import numpy as np
import matplotlib.pyplot as plt

def fnf(x):
    return np.sqrt(1 + x ** 3)

# Riemann sum 
def riemann_sum(f, a, b, n, method='left'):
    delta_x = (b - a) / n
    x_values = []
    heights = []
    
    if method == 'left':
        x_points = np.linspace(a, b - delta_x, n)
    elif method == 'right':
        x_points = np.linspace(a + delta_x, b, n)
    elif method == 'midpoint':
        x_points = np.linspace(a + delta_x / 2, b - delta_x / 2, n)
    else:
        raise ValueError("Method must be 'left', 'right', or 'midpoint'")
    
    for x in x_points:
        x_values.append(x)
        heights.append(f(x))
    
    sum_result = np.sum(np.array(heights) * delta_x)
    return sum_result, x_values, heights, delta_x

# Plotting 
def plot_riemann(f, a, b, n, method='left', show_rectangles=True):
    x = np.linspace(a, b, 1000)
    y = f(x)  
    plt.plot(x, y, label='f(x) = sqrt(1 + x^3)', color='blue')
    sum_result, x_values, heights, delta_x = riemann_sum(f, a, b, n, method)
    
# Draw rectangles if toggled on
    if show_rectangles:
        for x, height in zip(x_values, heights):
            plt.bar(x, height, width=delta_x, align='edge', alpha=0.3, edgecolor='black', color='cyan')
            plt.text(x + delta_x / 2, height / 2, f'{height:.2f}', ha='center', va='center', color='black')
            plt.text(x + delta_x / 2, -0.1, f'{delta_x:.2f}', ha='center', color='red')
    
    plt.title(f"Riemann Sum ({method.capitalize()} Endpoint) with n={n}")
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.figtext(0.15, 0.85, f'Riemann sum ({method} endpoint) approximation: {sum_result:.4f}', color='purple')
    plt.show()
    print(f"Riemann sum ({method} endpoint) approximation: {sum_result:.4f}")

a = 1  
b = 3  
n = 4  
plot_riemann(fnf, a, b, n, method='left', show_rectangles=True)
plot_riemann(fnf, a, b, n, method='right', show_rectangles=True)
plot_riemann(fnf, a, b, n, method='midpoint', show_rectangles=True)
```

```python
def plot_hero_graphic(f, a, b, n):
    x = np.linspace(a, b, 1000)
    y = f(x)
    
    plt.plot(x, y, label='f(x) = sqrt(1 + x^3)', color='black')
    
    # Colors for each 
    colors = {'left': 'cyan', 'right': 'orange', 'midpoint': 'purple'}
    methods = ['left', 'right', 'midpoint']
    
    for method in methods:
        # Calculate Riemann sum and get rectangle data
        _, x_values, heights, delta_x = riemann_sum(f, a, b, n, method)
        
        # Draw rectangles for each method with colors
        for x, height in zip(x_values, heights):
            plt.bar(x, height, width=delta_x, align='edge', alpha=0.4, edgecolor='black', color=colors[method], label=method.capitalize() + " Sum" if x == x_values[0] else "")
    
    plt.title(f"Comparison of Left, Right, and Midpoint Riemann Sums with n={n}")
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.show()
    
a = 1 
b = 3  
n = 4
plot_hero_graphic(fnf, a, b, n)
```

I made a small change to where the program picks the x-values for sampling. Originally, the program sampled at the left endpoint of each subinterval by starting at x = a. To sample at the midpoint instead, I adjusted the starting x-value to x = a + delta_x / 2, where delta_x is the width of each subinterval. This shift moves each sampling point to the middle of its interval.

Then, I updated the line that calculates the x-values for sampling, replacing it with x_values = np.linspace(a + delta_x / 2, b - delta_x / 2, n). This new range gives the midpoints of each subinterval rather than the left endpoints. With this simple change, the program now calculates the Riemann sum by sampling at the midpoints, which can give a more accurate approximation than sampling at the left or right endpoints.

```python
def fnf(x):
    return np.sqrt(1 + x ** 3)

def trapezoid_rule(func, a, b, number_of_steps):
    delta_x = (b - a) / number_of_steps
    x = a
    accumulation = 0
    
    for k in range(1, number_of_steps + 1):
        next_x = x + delta_x
        deltaS = (func(x) + func(next_x)) / 2 * delta_x
        accumulation += deltaS
        x = next_x
        print(f"Step {k}, deltaS: {deltaS:.10f}, Accumulation: {accumulation:.10f}")
    
    return accumulation

a = 1
b = 3
number_of_steps = 4

trapezoid_sum = trapezoid_rule(fnf, a, b, number_of_steps)
print(f"\nTrapezoid Rule Sum: {trapezoid_sum:.10f}")
```

```python
def simpsons_rule(func, a, b, number_of_steps):
    if number_of_steps % 2 != 0:
        raise ValueError("Simpson's Rule requires an even number of intervals.")
    
    delta_x = (b - a) / number_of_steps
    accumulation = func(a) + func(b)  # Start with f(a) + f(b)

    for k in range(1, number_of_steps):
        x = a + k * delta_x
        if k % 2 == 1:
            accumulation += 4 * func(x)
        else:
            accumulation += 2 * func(x)
        print(f"Step {k}, x: {x:.2f}, Current Term: {func(x):.10f}, Accumulation: {accumulation:.10f}")
    
    accumulation = (delta_x / 3) * accumulation
    return accumulation

a = 1
b = 3
number_of_steps = 4  

simpsons_sum = simpsons_rule(fnf, a, b, number_of_steps)
print(f"\nSimpson's Rule Sum: {simpsons_sum:.10f}")
```

The Trapezoid Rule and Simpsons Rule both offer significant advantages over basic Riemann sums when calculating approximate integrals. With the Trapezoid Rule, I gain better accuracy because it averages the function values at each intervals endpoints, creating a trapezoid instead of a rectangle to approximate the area. This small adjustment brings the approximation closer to the true shape of the curve, reducing the error, especially for functions with gentle slopes. Simpsons Rule goes even further by using parabolic segments instead of straight lines, which can provide remarkably high accuracy with fewer subintervals. This is especially useful for smooth or quadratic-like functions, as the parabolic segments capture the curves shape more accurately. In practice, this means I can achieve a higher degree of precision with fewer calculations, making Simpsons Rule both accurate and efficient.
